```python
import yfinance as yf
import pandas as pd 
import matplotlib.pyplot as plt

#Define stock symbol and time period
stock_symbol = 'TSLA'
start_date = '2018-01-01'
end_date = '2023-07-27'

#using yahoo finance for this data analysis
stock_data = yf.download(stock_symbol, start=start_date, end=end_date)

#Calculate moving averages and daily returns
stock_data['SMA_50']=stock_data['Close'].rolling(window=50).mean()
stock_data['SMA_200']=stock_data['Close'].rolling(window=200).mean()
stock_data['Daily_Return']=stock_data['Close'].pct_change()

#plt using matplotlib
plt.figure(figsize=(10,6))
plt.plot(stock_data.index, stock_data['Close'], label='Stock Price')
plt.plot(stock_data.index, stock_data['SMA_50'], label='SMA 50 days')
plt.plot(stock_data.index, stock_data['SMA_200'], label='SMA 200 days')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title(f'{stock_symbol} Stock Price with SMAs')
plt.legend()
plt.show()
```
